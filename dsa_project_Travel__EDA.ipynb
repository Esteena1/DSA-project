3. Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.displot(df.MonthlyIncome,color='r')
plt.title('MonthlyIncome',size=18)
plt.xlabel('Income',size=14)
plt.ylabel('Density',size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df.Age)
plt.title('Age Distribution',size=18)
plt.xlabel('Age',size=14)
plt.ylabel('Count',size=14)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(df.MonthlyIncome)
plt.title('Monthly Income',size=18)
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.countplot(data=df, x='CityTier', palette='Set2')
plt.title('City Tier Distribution',size=18)
plt.xlabel('City',size=14)
plt.ylabel('Count',size=14)
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.countplot(data=df, x='Passport', palette='Set2')
plt.title('Passport Distribution',size=18)
plt.xlabel('Passport',size=14)
plt.ylabel('Count',size=14)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x='Age',y='MonthlyIncome',hue='CityTier',color='r',data=df)
plt.title('Age vs Monthly Income',size=18)
plt.xlabel('Age',size=14)
plt.ylabel('Monthly Income',size=14)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.set_style('darkgrid')
sns.boxplot(x='CityTier',y='MonthlyIncome',data=df)
plt.title('City v/s Monthly Income',size=18);
```

```python
sns.pairplot(df,
                 markers="+",
                 diag_kind="kde",
                 kind='reg',
                 plot_kws={'line_kws':{'color':'#aec6cf'},
                           'scatter_kws': {'alpha': 0.7,
                                           'color': 'red'}},
                 corner=True);
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(df.corr(),annot=True,square=True,
            cmap='RdBu',
            vmax=1,
            vmin=-1)
plt.title('Correlations Between Variables',size=18);
plt.xticks(size=13)
plt.yticks(size=13)
plt.show()
```
